### Year

```python
kobe['year'] = kobe['game_date'].dt.year
```

```python
kobe['year'].describe()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='year', hue='shot_made_flag', order=range(1996,2017), data=kobe)
```

```python
count = pd.DataFrame(kobe.groupby('year').count()['shot_made_flag'])
count.rename(columns={'shot_made_flag':'total'}, inplace=True)
summ = pd.DataFrame(kobe.groupby('year').sum()['shot_made_flag'])
summ.rename(columns={'shot_made_flag':'shot_made'}, inplace=True)
prop_table = pd.concat([count,summ], axis=1)
prop_table['proportion'] = prop_table['shot_made']/prop_table['total']
prop_table['year'] = prop_table.index
prop_table
```

```python
plt.figure(figsize=(15,10))
sns.pointplot(x='year', y='proportion', data=prop_table)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.title('Shot Performance by Year', fontsize=20)
```

### Days

```python
count = pd.DataFrame(kobe.groupby('game_date').count()['shot_made_flag'])
count.rename(columns={'shot_made_flag':'total'}, inplace=True)
summ = pd.DataFrame(kobe.groupby('game_date').sum()['shot_made_flag'])
summ.rename(columns={'shot_made_flag':'shot_made'}, inplace=True)
prop_table = pd.concat([count,summ], axis=1)
prop_table['proportion'] = prop_table['shot_made']/prop_table['total']
prop_table['date'] = prop_table.index
```

```python
plt.figure(figsize=(15,10))
sns.pointplot(x='date', y='proportion', data=prop_table)
plt.xlabel('date', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.title('Shot Performance by Day', fontsize=20)
```

```python
plt.figure(figsize=(15,10))
plt.plot(prop_table['proportion'], alpha=0.5)
plt.xlabel('Day', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.title('Shot Performance by Day', fontsize=20)
```

### Months

First using the "day" data to calculate running average and median over 30 day intervals

```python
# potentially something happening by month?
n = 30
x = range(1,len(prop_table)+2)
y = np.append(prop_table['proportion'].values, (np.random.uniform(size=1)[0]))
running_avg = np.mean(y.reshape(-1, n), axis=1)

plt.figure(figsize=(15,10))
plt.plot(running_avg[1:])
plt.xlabel('Day', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.title('Shot Performance - Running 30 Day Average', fontsize=20)
plt.show()
```

```python
# potentially something happening by month?
n = 30
x = range(1,len(prop_table)+2)
y = np.append(prop_table['proportion'].values, (np.random.uniform(size=1)[0]))
running_med = np.median(y.reshape(-1, n), axis=1)

plt.figure(figsize=(15,10))
plt.plot(running_avg[1:], label='Mean')
plt.plot(running_med[1:], label='Median')
plt.xlabel('Day', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.legend(loc=1,prop={'size':20})
plt.title('Shot Performance - Running 30 Day Mean and Median', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='month', hue='shot_made_flag', order=range(1,13), data=kobe)
```

```python
count = pd.DataFrame(kobe.groupby('month').count()['shot_made_flag'])
count.rename(columns={'shot_made_flag':'total'}, inplace=True)
summ = pd.DataFrame(kobe.groupby('month').sum()['shot_made_flag'])
summ.rename(columns={'shot_made_flag':'shot_made'}, inplace=True)
prop_table = pd.concat([count,summ], axis=1)
prop_table['proportion'] = prop_table['shot_made']/prop_table['total']
prop_table['month'] = prop_table.index

plt.figure(figsize=(15,10))
sns.pointplot(x='month', y='proportion', data=prop_table)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.title('Shot Performance by Month', fontsize=20)
```

Lots of variability here. I wonder if it's mostly affected by shot volume, though?

```python
prop_table['shot_count'] = kobe.groupby('month').count()['action_type']
prop_table['shot_count_prop'] = prop_table['shot_count']/np.sum(prop_table['shot_count'])
```

```python
plt.figure(figsize=(15,10))
sns.pointplot(x='month', y='shot_count', data=prop_table)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Proportion of Shots Taken', fontsize=15)
plt.title('Shots Taken Per Month', fontsize=20)
```

```python
plt.figure(figsize=(15,10))
sns.pointplot(x='month', y='proportion', color='r', label='Prop', data=prop_table)
sns.pointplot(x='month', y='shot_count_prop', label='Total', data=prop_table)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Proportion of Shots Made', fontsize=15)
plt.title('Shot Performance by Month', fontsize=20)
#plt.legend(frameon=True)
```

Hard to tell from the plot. His shot proportion could be a reflection of shots taken but this isn't necessarily the case. For example, he took the least amount of shots in October, but this doesn't represent his worst shooting month. Still, the scale is difficult to compare.

# Categorical variable preprocessing

```python
kobe.head()
```

```python
le = preprocessing.LabelEncoder()
le.fit(kobe['shot_zone_area'])
kobe['sza'] = le.transform(kobe['shot_zone_area'])
```

```python
kobe[['sza','shot_zone_area']].head()
```

```python
# action type
kobe['at'] = le.fit_transform(kobe['action_type'])

# combined shot type
kobe['cst'] = le.fit_transform(kobe['combined_shot_type'])

# shot type
kobe['st'] = le.fit_transform(kobe['shot_type'])

# shot zone basic
kobe['szb'] = le.fit_transform(kobe['shot_zone_basic'])

# shot zone range
kobe['szr'] = le.fit_transform(kobe['shot_zone_range'])

# opponent
kobe['opp'] = le.fit_transform(kobe['opponent'])

kobe[['at','action_type','cst','combined_shot_type','st','shot_type','szb','shot_zone_basic','szr','shot_zone_range','opp','opponent']]
```
